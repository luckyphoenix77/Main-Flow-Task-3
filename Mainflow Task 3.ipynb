```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv("C:\\Users\\harsh\\Downloads\\Household-living-costs-price-indexes-march-2024-quarter-group-facts.csv")
```

```python
display(data.head(10))
```

```python
#Scatter plot against year and own
plt.scatter(data['year'],data['own'])
plt.title("Scatter Plot")
plt.xlabel("Year")
plt.ylabel("Own")
plt.show()
```

```python
#Line chart against year and own
plt.plot(data['year'],color = "cyan")
plt.plot(data['own'],color = "orange")
plt.title("Line chart")
plt.xlabel("Year")
plt.ylabel("Own")
plt.show()
```

```python
#Bar chart
plt.bar(data['year'],data['own'],color="gray")
plt.title("Bar chart")
plt.xlabel("Year")
plt.ylabel("Own")
plt.show()
```

```python
#Histogram
plt.hist(data['income'])
plt.title("Histogram")
plt.show()
```
